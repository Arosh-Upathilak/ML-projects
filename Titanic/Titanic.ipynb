### Handing missing value

```python
titanic_data = titanic_data.drop(columns='Cabin', axis=1) #drop Cabin folder
```

```python
titanic_data.head(10)
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True) # replacing the missing values in "Age" column with mean value
```

```python
titanic_data.isnull().sum()
```

```python
print(titanic_data['Embarked'].mode()) #most use Embarked is 'S'
```

```python
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True) # replacing the missing values in "Embarked" column with mode value
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.shape
```

```python
titanic_data.describe()
```

```python
titanic_data['Survived'].value_counts()
```

```python
df = pd.DataFrame(titanic_data) #A DataFrame is a core data structure in pandas, designed for working with tabular data.
numeric_columns = df.select_dtypes(include=['number'])
sns.heatmap(numeric_columns.corr(), cmap='YlGnBu', annot=True)
plt.show()

#only remove in the near of 0 aothers are consider
```

```python
titanic_data.head(10)
```

```python
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
#dummy_Sex=pd.get_dummies(titanic_data['Sex'],drop_first=True,prefix='Sex')
#dummy_Sex=dummy_Sex.astype(int)
#titanic_data.drop(columns=['Sex'],inplace=True)
#titanic_data=pd.concat([titanic_data,dummy_Sex],axis=1)
```

```python
X= titanic_data.drop(columns=['PassengerId','Name','Ticket','Survived'],axis=1)
#titanic_data.drop(columns=['Sex'],inplace=True) must delete colunm
Y=titanic_data['Survived']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Traning

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
X_test_prediction=model.predict(X_test)
```

```python
X_test_prediction
```

```python
Y_test
```

```python
print('Accuracy score of test data : ', accuracy_score(Y_test,X_test_prediction))
```
